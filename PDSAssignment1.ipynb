```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/datasets/hepatitis/master/data/hepatitis.csv')
df.head(10)
```

```python
df.shape
```

```python
numeric_df = df.select_dtypes(include=np.number).copy()
numeric_df
```

```python
# finding the missing values
missingvalues = df[df.isnull().any(axis=1)]
print(missingvalues)
```

```python
# Calculate the mean of only the numeric columns
mean_values = df.select_dtypes(include=np.number).mean()
print(mean_values)
```

```python
# Calculate the mean of each numeric column and replace missing values with the mean
for col in df.select_dtypes(include=np.number).columns:
  # Calculate the mean of the column
  mean_val = df[col].mean()
  # Replace missing values in the column with the mean
  df[col].fillna(mean_val, inplace=True)
```

```python
numeric_cols = df.select_dtypes(include=['number'])

# Check if there are any remaining missing values in numeric columns
missing_values_count = numeric_cols.isnull().sum()
print(missing_values_count)
```

```python
#the missing values are replace and we can see only the missing values for categorical data only now
missing_values_count = df.isnull().sum()
print(missing_values_count)
```
